####2.3. Quick tour of combinations

For an introduction or a refresher on combinations, please see: https://en.wikipedia.org/wiki/Combination 

$nCr$ is read as "n choose r" and can also be denoted as $n \choose r$, $C(n,r)$, $C^{n}_{r}$, etc.



$nCr$ is a function that outputs the number of conceivable ways for choosing $r$ discrete options from available $n$ discrete options. 

Here:  
 -  Here [discritization](https://en.wikipedia.org/wiki/Discretization) implies that $n , r \in  (0, 1, 2, 3, ...)$

 - Since choosing less than none of the available options is inconceivable $r \ge 0$ 

 - Since choosing more than all of the available options is inconceivable $r \le n$. 
So we can rewrite this as $\space n - r \ge 0$

 - As a shorter expression: $\space n \ge r \ge 0$  
 
 - Also,  $r$ options are chosen arbitrarily from the available $n$ options. The arrangement of or adjacency amongst the $r$ chosen options is inconsequential when constructing combinations of $r$ from $n$. 

###3. Factorials

####3.1. Defining a factorial

Let $x \in \mathbb{N}_0$ 

The notation $x!$ is read as "$x$ factorial"

Let $x!$ be the product of all natural numbers in the closed interval $[1, x]$ 

- The expression in desscending arrangement of terms:  
$x! = x(x-1)(x-2)(x-3) ... (x-(x-1))$  

  Example:  
  $\quad 5! = (5)(4)(3)(2)(1)$  

- The expression in asscending arrangment of terms:  
$x! = (x - (x-1))(x - (x - 2)) (x - (x-3))... (x)$  

  Example:  
  $\quad 5! = (1)(2)(3)(4)(5)$  
  >
  $\quad \text{eval}(5!) = 120$


#####Note

- For more on factorials and factorial notation see: https://en.wikipedia.org/wiki/Factorial   

- Regarding interval notation see: https://en.wikipedia.org/wiki/Interval_(mathematics)

- $eval(Q)$ is a function that evaluates an expression $Q$ using  gramatical rules constituting the definition of $eval()$

- We can also say that $x!$ is the production of each countable magnitude upto and including $x$ from a datum as a geometric progression within an [Abelian Group](https://en.wikipedia.org/wiki/Abelian_group)

####3.2. Observations about $x!$


1. $ x! = x(x-1)!$  
1. $0!$ is the product of natural numbers in the span from $1$ to $0$ with $1$ and $0$ included in the span   
  - **This does not mean that** $0! = (1)(0) = (0)(1) = 0$  
  - Since only the natural number found in $[1,0]$ is $1$, $\space 0! = 1$
1. Similarly $1! \in [1,1]\space$ so, $1!= 1$

####3.3. Review

$\quad x \in \mathbb{N}_0$

$\quad x! \in [1,x]\space$ where $[1,x] \in \mathbb{N}_1$   

$\quad \text{eval}(x!) = \text{c}\space$ where $\text{c} \in \mathbb{N}_1$

####3.4. Counting the number of terms in $x!$

The notation $|C|$ is read as "modulus of $C$" or as "magnitude of $C$"

Here $C$ is an unordered list or collection of discrete countable terms. $C$ may be constituted by zero number of terms. 

Let $|C|$ is a function that outputs the number of terms in $C$. 

We note that $\space |C|\in \mathbb{N}$


So $|x!|$ is a function that outputs the number of terms in $x!$  where $|x!| \in \mathbb{N}$  
By using $x \in \mathbb{N}_0$ i.e. $x$ takes the values of $(0, 1, 2, ...)$ we find that   
$|x!| = 
\begin{cases}
    1       & \quad \text{if } x = 0 \\
    x  & \quad \text{if } x > 0 
\end{cases}
$  

Example:  
$\quad |6!| = 6$  
$\quad |(258 - 139 + 119)!| = 1$

####3.5. Index of a term in $(x!)$ 


The notation $[S ]_i$ is read as "index of $S$ at $i$".

Here $S$ is a list or a dictionary containing distinct countable terms ordered in a sequence and $\space |S| \ge i \ge 0$

Let $[S ]_i$ be a function that outputs the $i_{th}$ term in the sequence $S$   



Read the notation $(x!)$ as "sequenced $x$ factorial"

Let $\space (x!) := (x(x-1)(x-2)(x-3) ... (x-(x-1)))$  

So $[(x!)]_i$ is a function that outputs the $m_{i}$ term of $(x!)$

Where, $\space i, x \in \mathbb{N}_0$ and $\space x \ge i \ge 0 \space$  

$Ind(x!)_0 := m_0 := x$

$Ind(x!)_1 := m_1 := x-1$  

$Ind(x!)_2 := m_2 := x-2$  

$Ind(x!)_3 := m_3 := x-3$  
.  
.  
.

$Ind(x!)_i := m_i := x - i$  

The last term in $x!$ is $Ind(x!)_{x-1} := m_{x-1} := x-(x-1) = 1$

####5.2. Formula for $nCr$

$$nCr = \frac{n!} {r!  (n-r)!}$$

####5.3. Evolved example about  choosing

````
 #@title Choose r options from n options { run: "auto" }
#@markdown Let n = 5
n1 = False #@param {type:"boolean"}
n2 = False #@param {type:"boolean"}
n3 = False #@param {type:"boolean"}
n4 = False #@param {type:"boolean"}
n5 = False #@param {type:"boolean"}

from math import factorial
from itertools import combinations

def pascal_rule(n,r):
  if r == 0:
    n_1Cr_1 = 0
    n_1Cr = 1
    
    print('n = {}, r = {}, n_1Cr_1 = {}'.format(n, r, n_1Cr_1))
    print('n = {}, r = {}, n_1Cr = {}'.format(n, r, n_1Cr))
        
  while n-1-r >= 0 and r-1 >= 0:
    n_1Cr_1 = int(factorial(n-1)/
                  (factorial(r-1)*factorial(n-1-(r-1))))
    
    n_1Cr = int(factorial(n-1)/
                (factorial(r)*factorial(n-1-r)))
    
    print('n = {}, r = {}, n_1Cr_1 = {}'.format(n, r, n_1Cr_1))
    print('n = {}, r = {}, n_1Cr = {}'.format(n, r, n_1Cr))
    
    n -= 1
    r -= 1   


options = [n1, n2, n3, n4, n5]

opt_str = ['n1', 'n2', 'n3', 'n4', 'n5']

n = len(options)

chosen = list()
for i, n_val in enumerate(options):  
  if n_val is True:
    chosen.append(opt_str[i])
    
r = len(chosen)

comb_list = combinations(opt_str,r)

chosen_string = str()
for k in chosen: 
  chosen_string += k + ', ' 


pascal_rule(n,r)
  
# pascal_list = [item for item in pascal_rule(n,r)]

# print(pascal_list)
print('\n')
  
if r > 0 and r != n:
  nCr = factorial(n)/(factorial(r)*factorial(n-r))
  
else:
  nCr = 1

print('n = {}, r = {}'.format(n,r))

print('Chosen option(s): {} \n'
      .format(chosen_string.rstrip(', ')))

print('Total ways in which {} of {} options can be chosen: {}'
      .format(r,n,int(nCr)))

if r == 0:
  print('\nThere is exactly one way of choosing none of the options!')

if r == n:
  print('\nThere is exactly one way of choosing all of the options!')

comb_str = str()
for j, comb in enumerate(comb_list):

  comb_str = ', '.join(comb)

  print('#{:^2}: {}'.format(j+1,comb_str))

#     if r != n:
#       for row_val in pascal_list:
        
  
# print(comb_list)
````
